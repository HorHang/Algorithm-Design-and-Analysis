```python
red = 'red'
black = 'black'

class TreeNode:
    def __init__(self,key):
        self.key = key
        self.left = None
        self.right = None
        self.p = None
        self.color = red

class RedBlackTree():
    def __init__(self):
        nil_node = TreeNode(0)
        nil_node.color = black
        self.nil = nil_node
        self.root = self.nil

    def left_rotation(self,x):
        y = x.right
        x.right = y.left
        if y.left is not self.nil:
            y.left.p = x
        y.p = x.p
        if x.p is self.nil:
            self.root = y
        elif x is x.p.left:
            x.p.left = y
        else:
            x.p.right = y
        y.left = x          # x become y's left child
        x.p = y        # y become x's parent

    def right_rotation(self,y):
        x = y.left
        y.left = x.right
        if x.right is not self.nil:
            x.right.p = y
        x.p = y.p
        if y.p is self.nil:
            self.root = x
        elif y is y.p.right:
            y.p.right = x
        else:
            y.p.left = x
        x.right = y
        y.p = x

    def search(self,z):
        x = self.root
        while x != self.nil and x.key != z.key:
            if x.key < z.key:
                x = x.right
            else:
                x = x.left
        return x

    def find_min(self,x):
        while x.left is not self.nil:
            x = x.left
        return x

    def find_max(self,x):
        while x.right is not self.nil:
            x = x.right
        return x

#### Without using comparison
    def successor(self,x):
        if x.right != self.nil:   
            return self.find_min(x.right)                  # find min of x's right subtree
        y = x.p
        while y is not self.nil and x is y.right:       # x is right child
            x = y
            y = y.p
        return y

    def predecessor(self,x):
        if x.left != self.nil:
            return self.find_max(x.left)
        y = x.p
        while y is not self.nil and x is y.left:
            x = y
            y = y.p
        return y

    def transplant(self,u,v):               # Replace subtree u by subtree v
        if u.p is self.nil:                # u is root
            self.root = v
        elif u is u.p.right:             # u is right subtree
            u.p.right = v
        else:                               # u is left subtree
            u.p.left = v
        v.p = u.p

    def inorder(self,tree):
        if tree is not self.nil:
            self.inorder(tree.left)
            print(tree.color, end=" ")
            self.inorder(tree.right)
    def insert(self,z):
    # Initialization
        y = self.nil
        x = self.root
    # Tree Traversal to find the parent of target insert node
        while x != self.nil:
            y = x
            if z.key < x.key:
                x = x.left
            else:
                x = x.right
        z.p = y
    # empty tree
        if y is self.nil:
            self.root = z
    # Assign the node z to be left or right child
        elif z.key < y.key:
            y.left = z
        else:
            y.right = z
        z.left = self.nil
        z.right = self.nil
        # z.color = red
        self.RB_Insert_Fixup(z)

    def RB_Insert_Fixup(self,z):
        while z.p.color is red:
            if z.p is z.p.p.left:
                y = z.p.p.right
                if y.color is red:
                    z.p.p.color = red
                    y.color = black
                    z.p.color = black
                    z = z.p.p
                else: #y.color is black
                    if z is z.p.right:
                        z = z.p
                        self.left_rotation(z)
                    z.p.color = black
                    z.p.p.color = red
                    self.right_rotation(z.p.p)
            else: #z.p is z.p.p.right
                y = z.p.p.left
                if y.color is red:
                    z.p.p.color = red
                    y.color = black
                    z.p.color = black
                    z = z.p.p
                else:
                    if z is z.p.left:
                        z = z.p
                        self.right_rotation(z)
                    z.p.color = black
                    z.p.p.color = red
                    self.left_rotation(z.p.p)
        self.root.color = black

    def delete(self,z):
        y = z
        y_original_color = y.color
        if z.left is self.nil:               # Has right subtree
            x = z.right
            self.transplant(z,z.right)
        elif z.right is self.nil:            # Has left subtree
            x = z.left
            self.transplant(z,z.left)
        else:                               # Has both left & right subtree
            y = self.find_min(z.right)
            y_original_color = y.color
            x = y.right
            if y.p != z:
                self.transplant(y,y.right)
                y.right = z.right
                y.right.p = y
            self.transplant(z,y)
            y.left = z.left
            y.left.p = y
            y.color = z.color
        if y_original_color is black:
            self.RB_Delete_Fixup(x)

    def RB_Delete_Fixup(self,x):
        while x is not self.nil and x.color is black:
            if x is x.p.left:
                w = x.p.right # w is x's sibling
                if w.color is red:
                    w.color = black
                    x.p.color = red
                    self.left_rotation(x.p)
                    w = x.p.right
                if w.left.color is black and w.right.color is black:
                    w.color = red
                    x = x.p
                else:
                    if w.right.color is black:
                        w.color = red
                        w.left.color = black
                        self.right_rotation(w)
                        w = x.p.right
                    w.color = x.parent.color
                    x.p.color = black
                    w.right.color = black
                    self.left_rotation(x.p)
                    x = self.root
            else: # x is x.p.right
                w = x.p.left # w is x's sibling
                if w.color is red:
                    w.color = black
                    x.p.color = red
                    self.right_rotation(x.p)
                    w = x.p.left
                if w.right.color is black and w.left.color is black:
                    w.color = red
                    x = x.p
                else:
                    if w.left.color is black:
                        w.color = red
                        w.right.color = black
                        self.left_rotation(w)
                        w = x.p.left
                    w.color = x.parent.color
                    x.p.color = black
                    w.left.color = black
                    self.right_rotation(x.p)
                    x = self.root
        x.color = black

            # if y.p is y.p.p.right:
            #     x = y.left
            #     w = y.p.right
            #     if w.color is red:
            #         self.left_rotation(w)
            #         w = x.p.right
            #     if w.left.color is black and w.right.color is black:
            #         w.color = red
            #         x = x.p
            #     elif w.left.color is red:
            #         self.right_rotation(w)
            #         w.color = red
            #         w = w.p
            #     else:
            #         x.p.color = black
            #         w.color = red
            #         self.left_rotation(w)
            #         x = self.root

tree = RedBlackTree()
a = TreeNode(7)
b = TreeNode(4)
c = TreeNode(11)
d = TreeNode(3)
e = TreeNode(6)
f = TreeNode(9)
i = TreeNode(18)
j = TreeNode(14)
k = TreeNode(19)
l = TreeNode(12)
m = TreeNode(17)
n = TreeNode(22)
o = TreeNode(20)

tree.insert(a)
tree.insert(b)
tree.insert(c)
tree.insert(d)
tree.insert(e)
tree.insert(f)
tree.insert(i)
tree.insert(j)
tree.insert(k)
tree.insert(l)
tree.insert(m)
tree.insert(n)
tree.insert(o)

tree.delete(n)
tree.inorder(tree.root)
```
